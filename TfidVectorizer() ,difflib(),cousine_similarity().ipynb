## .TfidVectorizer()

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data
documents = ["This is the first document.",
             "This document is the second document.",
            "And this is the third one.",
             "Is this the first document?"]

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(documents)

# Get the TF-IDF features
feature_names = vectorizer.get_feature_names_out()

# Convert the TF-IDF matrix to a Pandas DataFrame for easier visualization
import pandas as pd
df = pd.DataFrame(data=tfidf_matrix.toarray(), columns=feature_names)
df
```

## .difflib

```python
import difflib

# List of words
word_list = ["apple", "banana", "cherry", "date", "fig", "grape", "lemon", "lime"]

# Find close matches for a misspelled word
misspelled_word = "leomn"
close_matches = difflib.get_close_matches(misspelled_word, word_list)

print(f"Close matches for '{misspelled_word}': {close_matches}")
```
